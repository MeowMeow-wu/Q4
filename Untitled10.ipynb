## Q4 (a) Explain the problem and why AI is suitable.

### The Problem:
CompanyA, a charity organization, faces the challenge of accurately predicting the daily demand for lunchboxes over a 30-day period, coinciding with the Hari Raya festive season. They have a fixed target of providing lunchboxes to 40 families. The core problem is to determine how many lunchboxes will be needed each day to meet the needs of these families without excessive waste or shortages. Factors influencing this demand can include:
* The specific needs or availability of the 40 families on any given day.
* The impact of the Hari Raya period on family routines and needs.
* Potential external factors like weather, local events, or logistical challenges.

An incorrect prediction can lead to either:
* **Food Waste:** Preparing too many lunchboxes, leading to financial and resource inefficiency for the charity.
* **Unmet Needs:** Preparing too few lunchboxes, failing to serve all the intended families and potentially causing disappointment.

### Why AI is Suitable:
AI, particularly machine learning, is well-suited for this problem due to its ability to:
* **Identify Complex Patterns:** Food demand is rarely static. It can be influenced by subtle, interconnected factors (e.g., day of the week, proximity to Hari Raya, weather). AI algorithms can analyze historical data to uncover these non-linear relationships and patterns that might be missed by traditional forecasting methods or human intuition.
* **Handle Time-Series Data:** The problem involves predicting demand over a sequence of days, making it a time-series forecasting task. AI models are adept at learning from sequential data and identifying trends, seasonality, and holiday effects.
* **Optimize Resource Allocation:** By providing more accurate demand predictions, AI enables CompanyA to optimize the number of lunchboxes prepared. This directly translates to efficient use of their budget and resources, minimizing waste and ensuring that aid reaches those who need it most.
* **Adaptability:** AI models can be retrained with new data as it becomes available. This allows the prediction system to adapt to changing circumstances or evolving demand patterns, maintaining its accuracy over time.
* **Quantifiable Forecasting:** AI provides quantitative predictions (e.g., "predicting 35 lunchboxes for Tuesday"), which are actionable for planning procurement and preparation.

## 1. Data Collection (Simulated Historical Data)

````python
import pandas as pd
import numpy as np

# Simulate historical data for 1 year
start_date = '2023-01-01'
end_date = '2023-12-31'
dates = pd.date_range(start=start_date, end=end_date, freq='D')

# Simulate daily lunchbox demand
# Base demand + weekly seasonality + some random noise + a 'Hari Raya' like peak
demand = np.random.randint(25, 35, size=len(dates)) # Base demand

# Add weekly seasonality (e.g., lower on weekends, higher mid-week)
for i, date in enumerate(dates):
    if date.dayofweek < 5: # Weekday
        demand[i] += np.random.randint(0, 5)
    else: # Weekend
        demand[i] -= np.random.randint(0, 3)

# Simulate a 'Hari Raya' effect somewhere in the data
hari_raya_start = pd.to_datetime('2023-04-22') # Example date
hari_raya_end = pd.to_datetime('2023-05-05') # Example duration

hari_raya_period = (dates >= hari_raya_start) & (dates <= hari_raya_end)
demand[hari_raya_period] += np.random.randint(5, 15, size=hari_raya_period.sum()) # Increased demand during Hari Raya

# Ensure demand doesn't go below 0 (or a reasonable minimum)
demand = np.maximum(0, demand)

historical_df = pd.DataFrame({'Date': dates, 'Lunchboxes_Distributed': demand})

display(historical_df.head())
display(historical_df.tail())
````

## 2. Data Preprocessing and Feature Engineering

````python
historical_df['Date'] = pd.to_datetime(historical_df['Date'])

# Time-Based Features
historical_df['DayOfWeek'] = historical_df['Date'].dt.dayofweek # Monday=0, Sunday=6
historical_df['DayOfMonth'] = historical_df['Date'].dt.day
historical_df['Month'] = historical_df['Date'].dt.month
historical_df['Year'] = historical_df['Date'].dt.year
historical_df['WeekOfYear'] = historical_df['Date'].dt.isocalendar().week.astype(int)
historical_df['IsWeekend'] = historical_df['Date'].dt.dayofweek.isin([5, 6]).astype(int)

# Hari Raya Holiday Feature (using the simulated period)
hari_raya_start_forecast = pd.to_datetime('2024-04-10') # Example Hari Raya for next year's forecast
hari_raya_end_forecast = pd.to_datetime('2024-04-23') # Example duration for next year's forecast

# For historical data, we use the simulated Hari Raya period
historical_df['IsHariRayaPeriod'] = ((historical_df['Date'] >= hari_raya_start) &
                                    (historical_df['Date'] <= hari_raya_end)).astype(int)

# Lagged Features (simple example, more sophisticated would consider stationarity)
historical_df['Demand_Lag_1'] = historical_df['Lunchboxes_Distributed'].shift(1)
historical_df['Demand_Lag_7'] = historical_df['Lunchboxes_Distributed'].shift(7)

# Fill NaN values created by shifting (e.g., with the mean or 0, depending on context)
historical_df = historical_df.fillna(0) # For simplicity, fill with 0

display(historical_df.head(10))
````

## 3. Model Selection and Training (Facebook Prophet)

````python
# Install Prophet (if not already installed)
# !pip install prophet

import prophet
from prophet import Prophet

# Prepare data for Prophet: Prophet expects columns named 'ds' (datestamp) and 'y' (target variable)
prophet_df = historical_df.rename(columns={'Date': 'ds', 'Lunchboxes_Distributed': 'y'})

# Define custom holidays for Prophet (Hari Raya)
# Prophet expects a dataframe with 'holiday', 'ds', 'lower_window', 'upper_window' columns
holidays = pd.DataFrame({
    'holiday': 'hari_raya',
    'ds': pd.to_datetime([
        '2023-04-22', '2023-04-23', '2023-04-24', # Example dates for 2023 Hari Raya
        # Add other specific Hari Raya related dates that might affect demand
    ]),
    'lower_window': -2, # Start 2 days before the official date
    'upper_window': 5   # End 5 days after the official date
})

# Initialize and fit the Prophet model
# We'll include yearly, weekly, and daily seasonality by default.
# We can also add regressors for the engineered features.
model = Prophet(
    seasonality_mode='multiplicative', # Good for demand forecasting where seasonality scales with magnitude
    holidays=holidays,
    changepoint_prior_scale=0.05 # Adjust flexibility of trend
)

# Add custom regressors (engineered features)
# Note: Prophet only accepts numerical regressors.
model.add_regressor('DayOfWeek')
model.add_regressor('DayOfMonth')
model.add_regressor('Month')
model.add_regressor('Year')
model.add_regressor('WeekOfYear')
model.add_regressor('IsWeekend')
model.add_regressor('IsHariRayaPeriod')
model.add_regressor('Demand_Lag_1')
model.add_regressor('Demand_Lag_7')


model.fit(prophet_df)
print("Prophet model training complete.")
````

## 4. Prediction

````python
# Create a future DataFrame for the 30-day forecast period
future_period_start = pd.to_datetime('2024-04-01') # Start of 30-day forecast before Hari Raya
future_period_end = pd.to_datetime('2024-04-30') # End of 30-day forecast

future_dates = pd.date_range(start=future_period_start, end=future_period_end, freq='D')
future_df = pd.DataFrame({'ds': future_dates})

# Re-engineer features for the future DataFrame
future_df['DayOfWeek'] = future_df['ds'].dt.dayofweek
future_df['DayOfMonth'] = future_df['ds'].dt.day
future_df['Month'] = future_df['ds'].dt.month
future_df['Year'] = future_df['ds'].dt.year
future_df['WeekOfYear'] = future_df['ds'].dt.isocalendar().week.astype(int)
future_df['IsWeekend'] = future_df['ds'].dt.dayofweek.isin([5, 6]).astype(int)

# Hari Raya period for the forecast
future_df['IsHariRayaPeriod'] = ((future_df['ds'] >= hari_raya_start_forecast) &
                                 (future_df['ds'] <= hari_raya_end_forecast)).astype(int)

# For lagged features in the future, we need to extend the historical data or use previous predictions.
# For simplicity here, we'll use the last known values or zeros if out of range for the dummy data.
# In a real scenario, you'd use a more robust method, potentially iterating or using previous predictions.

# For demonstration, fill lagged features with the last known historical values or a sensible default.
# This is a simplification; a robust solution would handle this carefully.
last_known_demand = historical_df['Lunchboxes_Distributed'].iloc[-1]
last_known_demand_lag_7 = historical_df['Lunchboxes_Distributed'].iloc[-7]

future_df['Demand_Lag_1'] = np.nan # Will fill these more realistically later if needed
future_df['Demand_Lag_7'] = np.nan # Will fill these more realistically later if needed

# A simple way to handle initial lagged features for the forecast period:
# For the first few days of the forecast, use actual historical data.
# After that, use previous day's forecast as lag_1 and 7 days ago forecast as lag_7.
# This requires an iterative forecasting approach or a model that can handle future regressors.
# For now, let's fill with a placeholder for demonstration purposes.
future_df['Demand_Lag_1'] = last_known_demand # Simplification
future_df['Demand_Lag_7'] = last_known_demand_lag_7 # Simplification

# Generate predictions
forecast = model.predict(future_df)

# Display the forecast
display(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
display(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())

# Plot the forecast
fig = model.plot(forecast)
ax = fig.gca()
ax.set_title('Lunchbox Demand Forecast for 30 Days')
ax.set_xlabel('Date')
ax.set_ylabel('Predicted Lunchboxes')

# Plot components
fig_comp = model.plot_components(forecast)

````

## 5. Model Evaluation (on Historical Test Set)

To evaluate the model's performance on historical data, we will split our `prophet_df` into a training set and a test set. The model will be trained on the training set, and then we will make predictions on the test set and compare them against the actual values to calculate MAE and RMSE.

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Define the split point for training and testing
# For time series, it's crucial to split chronologically.
# Let's use the last 30 days of the historical data as the test set.
train_size = len(prophet_df) - 30
train_df = prophet_df.iloc[:train_size]
test_df = prophet_df.iloc[train_size:]

print(f"Training data: {train_df['ds'].min()} to {train_df['ds'].max()} ({len(train_df)} days)")
print(f"Testing data: {test_df['ds'].min()} to {test_df['ds'].max()} ({len(test_df)} days)")

# Re-initialize and train the Prophet model on the training data
# We'll use the same model configuration as before
model_eval = Prophet(
    seasonality_mode='multiplicative',
    holidays=holidays,
    changepoint_prior_scale=0.05
)

# Add the same custom regressors
model_eval.add_regressor('DayOfWeek')
model_eval.add_regressor('DayOfMonth')
model_eval.add_regressor('Month')
model_eval.add_regressor('Year')
model_eval.add_regressor('WeekOfYear')
model_eval.add_regressor('IsWeekend')
model_eval.add_regressor('IsHariRayaPeriod')
model_eval.add_regressor('Demand_Lag_1')
model_eval.add_regressor('Demand_Lag_7')

model_eval.fit(train_df)
print("Prophet model training complete on historical training data.")

# Create future dataframe for prediction on the test set
future_eval_df = test_df.drop(columns=['y']).copy() # Drop 'y' as we are predicting it

# Ensure all regressors are present in the future_eval_df
# The test_df already has these features since it's from the original historical_df

# Generate predictions for the test set
forecast_eval = model_eval.predict(future_eval_df)

# Extract actual and predicted values for the test period
actual_values = test_df['y'].values
predicted_values = forecast_eval['yhat'].values

# Calculate evaluation metrics
mae = mean_absolute_error(actual_values, predicted_values)
rmse = np.sqrt(mean_squared_error(actual_values, predicted_values))

print(f'\nMean Absolute Error (MAE) on historical test set: {mae:.2f}')
print(f'Root Mean Squared Error (RMSE) on historical test set: {rmse:.2f}')

# Optionally, visualize the test set prediction vs actuals
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(test_df['ds'], actual_values, label='Actual Demand', color='blue')
ax.plot(forecast_eval['ds'], predicted_values, label='Predicted Demand', color='red', linestyle='--')
ax.fill_between(forecast_eval['ds'], forecast_eval['yhat_lower'], forecast_eval['yhat_upper'], color='red', alpha=0.2, label='Prediction Interval')
ax.set_title('Actual vs. Predicted Lunchbox Demand on Test Set')
ax.set_xlabel('Date')
ax.set_ylabel('Lunchboxes')
ax.legend()
plt.grid(True)
plt.show()
````
